```python
#賣鴨子
Ln = int(input())
Lsum = 0
def fun(Ln,Lduck):
    Lsum = (Lduck + 1) * 2
    print(f'出發時共{Lsum}隻')
    print(f'賣{Lsum - Lduck}隻,剩{Lduck}隻')
    Ln = Ln - 1
    if Ln > 0 :
        fun(Ln,Lsum)

fun(Ln,2)
```

```python
def fun(Ln,Lint):
    La = (Lint + 1) * 2
    Ln -= 1
    print(f'出發時共趕了{La}隻鴨子')
    print(f'賣{La-Lint}隻,剩{Lint}隻')
    if Lint > 0 :
        fun(Ln,La)
Ln = 5

fun(Ln,2)
```

```python
#成績計算
Ln = int(input())
Lsum = [0,0,0]
for i in range(Ln):
    a = list(map(int,input().split()))
    for j in range(3):
        Lsum[j] += a[j]
    print(f'總分{sum(a)},平圴{(sum(a))//3}')
print(f'全班國文總分{Lsum[0]},平均{Lsum[0]//Ln}')
print(f'全班英文總分{Lsum[1]},平均{Lsum[1]//Ln}')
print(f'全班數學總分{Lsum[2]},平均{Lsum[2]//Ln}')

```

```python
Ln = int(input())
def pairs (xs):
  if 2 > len(xs):
    return []
  else:
    return [xs[0:2]] + pairs(xs[1:])

def pascal (n):
  def compute (prev):
    return [1] + [x + y for (x,y) in pairs(prev)] + [1]

  def aux (m, prev):
    if (m > n):
      return []
    else:
      return [prev] + aux(m + 1, compute(prev))
  return aux(1, [1])

[print(line) for line in pascal(Ln)]
```

```python
# 巴斯卡三角形
size = int(input())
dp = [[0] * (i + 1) for i in range(size)]  # 先把表建好
dp[0][0] = 1  # 第一格一定是1
for i in range(1, size):
    for j in range(i+1):
        if j == 0 or j == i:  # 第一個和最後一個都會是1
            dp[i][j] = 1
        else:
            dp[i][j] = dp[i-1][j-1] + dp[i-1][j]

for i in dp:
  print(' '.join(map(str,i)))
```
